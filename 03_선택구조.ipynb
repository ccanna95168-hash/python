    * `and` : ë‘ ì¡°ê±´ì´ ëª¨ë‘ ì°¸ì¼ ë•Œë§Œ ì°¸

```python
# ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„
import random

user = input('ê°€ìœ„, ë°”ìœ„, ë³´ ì¤‘ í•˜ë‚˜ë¥¼ ì„ íƒí•˜ì„¸ìš”: ')
com_rand = random.randint(1,3)

if com_rand == 1:
  computer = 'ê°€ìœ„'
elif com_rand == 2:
  computer = 'ë°”ìœ„'
else:
  computer = 'ë³´'

print(f'ì»´í“¨í„°: {computer}')

# ìŠ¹ë¶€
if user == computer:
  print('ë¬´ìŠ¹ë¬´')
elif user == 'ê°€ìœ„' and computer == 'ë³´' :
  print('ìŠ¹')
elif user == 'ë°”ìœ„' and computer == 'ê°€ìœ„' :
  print('ìŠ¹')
elif user == 'ë³´' and computer == 'ë°”ìœ„' :
  print('ìŠ¹')
else:
  print('íŒ¨')
```

## 4. í•¨ìˆ˜ (Function)
* **ê°œë…**: íŠ¹ì • ì‘ì—…ì„ ìˆ˜í–‰í•˜ëŠ” ëª…ë ¹ì–´ë“¤ì˜ ë¬¶ìŒì…ë‹ˆë‹¤.
* **ë¼ì´ë¸ŒëŸ¬ë¦¬ í™œìš©**: `import` í‚¤ì›Œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì™¸ë¶€ ë„êµ¬(ë¼ì´ë¸ŒëŸ¬ë¦¬)ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
def add(x,y):
  return x+y

def sub(x,y):
  return x-y

def mul(x,y):
  return x*y

def div(x,y):
  return x/y

a= int(input('ì •ìˆ˜1: '))
b= int(input('ì •ìˆ˜2: '))

print(f'ë”í•˜ê¸°:{add(a,b)}')
print(f'ë¹¼ê¸°:{sub(a,b)}')
print(f'ê³±í•˜ê¸°:{mul(a,b)}')
print(f'ë‚˜ëˆ„ê¸°:{div(a,b)}')
```

```python
# div í•¨ìˆ˜ì—ì„œ ëª« , ë‚˜ë¨¸ì§€ ë°˜í™˜
# ë³€ìˆ˜ ëª« Q , ë‚˜ë¨¸ì§€ R
# ëª«ê³¼ ë‚˜ë¨¸ì§€ ì¶œë ¥
def div(x,y):
  return x//y, x%y

a= int(input('ì •ìˆ˜1:'))
b= int(input('ì •ìˆ˜2:'))

Q,R = div(a,b)
print(f'ëª«:{Q}, ë‚˜ë¨¸ì§€:{R}')
```

### ğŸ“ ë‘ ë„í˜•ì˜ ë„“ì´ ë¹„êµí•˜ê¸°
1. **ê³„ì‚°**: ì‚¬ê°í˜•(ê°€ë¡œÃ—ì„¸ë¡œ)ê³¼ ì‚¼ê°í˜•(ë°‘ë³€Ã—ë†’ì´/2)ì˜ ì •ë³´ë¥¼ ê°ê° ì…ë ¥ë°›ì•„ ë„“ì´ë¥¼ êµ¬í•©ë‹ˆë‹¤.
2. **ë¹„êµ**:
    * `if ë„“ì´1 > ë„“ì´2:` ì²« ë²ˆì§¸ ë„í˜•ì´ ë” ë„“ë‹¤ê³  ì¶œë ¥í•©ë‹ˆë‹¤.
    * `elif ë„“ì´1 < ë„“ì´2:` ë‘ ë²ˆì§¸ ë„í˜•ì´ ë” ë„“ë‹¤ê³  ì¶œë ¥í•©ë‹ˆë‹¤.
    * `else:` ë‘ ë„í˜•ì˜ ë„“ì´ê°€ ê°™ë‹¤ê³  ì¶œë ¥í•©ë‹ˆë‹¤.

```python
# ì‚¬ê°í˜• square , ì‚¼ê°í˜• triangle, ì› circle í•¨ìˆ˜ 3ê°œ
# ì‚¬ê°í˜•ì€ ê°€ë¡œ,ì„¸í˜¸ ì…ë ¥ ë°›ì•„ì„œ ê°€ë¡œ*ì„¸ë¡œ ê°’ ë³€í™˜
# ì‚¼ê°í˜•ì€ ë°‘ë³€ base ë†’ì´height ì…ë ¥ ë°›ì•„ì„œ ì‚¼ê°í˜• ë„“ì´ëŠ” ë°‘ë³€*ë†’ì´/2 ê°’ ë³€í™˜
# ì›ì€ ë°˜ì§€ë¦„ radiusë¥¼ ì…ë ¥ ë°›ì•„ì„œ math.pi*ë°˜ì§€ë¦„*ë°˜ì§€ë¦„ ê°’ ë³€í™˜ (import.math ë¼ì´ë¸ŒëŸ¬ë¦¬ ì‚¬ìš©)
# ë„í˜•ì˜ ì¢…ë¥˜ë¥¼ ì…ë ¥ ë°›ì•„ì„œ ë³€ìˆ˜ shape_type1, shape_type2
# ì‚¬ê°í˜•ì´ë©´ ì‚¬ê°í˜• í•¨ìˆ˜ í˜¸ì¶œ, ì‚¼ê°í˜•ì´ë©´ ì‚¼ê°í˜• í•¨ìˆ˜ í˜¸ì¶œ, ì›ì´ë©´ ì› í•¨ìˆ˜ í˜¸ì¶œ
# area ë³€ìˆ˜ì—ëŠ” ê° ë„íš½ì˜ ë„“ì´ ê°’ì´ ì…ë ¥ë¨
# area ë³€ìˆ˜ì— ì…ë «ê´¸ ê°’, area1ê³¼ area2ë¥¼ ë¹„êµí•˜ì—¬ ì–´ë–¤ ë„í˜•ì´ ì–¼ë§ˆë§Œí¼ ë” ë„“ì€ì§€ ì¶œë ¥.

import math

def square(x,y):
  return x*y
def triangle(x,y):
  return x*y/2
def circle(radius):
  return radius*radius*math.pi

# ì²« ë²ˆì§¸ ë„í˜•ì˜ ë„“ì´ ê³„ì‚°
shape_type1 = input('ì²«ë²ˆì§¸ ë„í˜• ì¢…ë¥˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.(ì‚¬ê°í˜•,ì‚¼ê°í˜•,ì›)')
area1 = 0
if shape_type1 == 'ì‚¬ê°í˜•':
  height = int(input('ì²«ë²ˆì§¸ ì‚¬ê°í˜•ì˜ ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”: '))
  base = int(input('ì²«ë²ˆì§¸ ì‚¬ê°í˜•ì˜ ë°‘ë³€ì„ ì…ë ¥í•˜ì„¸ìš”: '))
  area1 = square(height,base)
  print(f'ì²«ë²ˆì§¸ ë„í˜•(ì‚¬ê°í˜•)ì˜ ë„“ì´: {area1}')
elif shape_type1 == 'ì‚¼ê°í˜•':
  height = int(input('ì²«ë²ˆì§¸ ì‚¼ê°í˜•ì˜ ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”: '))
  base = int(input('ì²«ë²ˆì§¸ ì‚¼ê°í˜•ì˜ ë°‘ë³€ì„ ì…ë ¥í•˜ì„¸ìš”: '))
  area1 = triangle(height,base)
  print(f'ì²«ë²ˆì§¸ ë„í˜•(ì‚¼ê°í˜•)ì˜ ë„“ì´: {area1}')
elif shape_type1 == 'ì›':
  radius = int(input('ì²«ë²ˆì§¸ ì›ì˜ ë°˜ì§€ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: '))
  area1 = circle(radius)
  print(f'ì²«ë²ˆì§¸ ë„í˜•(ì›)ì˜ ë„“ì´: {area1}')
else:
  print('ì˜ëª»ëœ ë„í˜• ì¢…ë¥˜ë¥¼ ì…ë ¥í–ˆìŠµë‹ˆë‹¤.')

# ë‘ ë²ˆì§¸ ë„í˜•ì˜ ë„“ì´ ê³„ì‚°
shape_type2 = input('ë‘ë²ˆì§¸ ë„í˜• ì¢…ë¥˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.(ì‚¬ê°í˜•,ì‚¼ê°í˜•,ì›)')
area2 = 0
if shape_type2 == 'ì‚¬ê°í˜•':
  height = int(input('ë‘ë²ˆì§¸ ì‚¬ê°í˜•ì˜ ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”: '))
  base = int(input('ë‘ë²ˆì§¸ ì‚¬ê°í˜•ì˜ ë°‘ë³€ì„ ì…ë ¥í•˜ì„¸ìš”: '))
  area2 = square(height,base)
  print(f'ë‘ë²ˆì§¸ ë„í˜•(ì‚¬ê°í˜•)ì˜ ë„“ì´: {area2}')
elif shape_type2 == 'ì‚¼ê°í˜•':
  height = int(input('ë‘ë²ˆì§¸ ì‚¼ê°í˜•ì˜ ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”: '))
  base = int(input('ë‘ë²ˆì§¸ ì‚¼ê°í˜•ì˜ ë°‘ë³€ì„ ì…ë ¥í•˜ì„¸ìš”: '))
  area2 = triangle(height,base)
  print(f'ë‘ë²ˆì§¸ ë„í˜•(ì‚¼ê°í˜•)ì˜ ë„“ì´: {area2}')
elif shape_type2 == 'ì›':
  radius = int(input('ë‘ë²ˆì§¸ ì›ì˜ ë°˜ì§€ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: '))
  area2 = circle(radius)
  print(f'ë‘ë²ˆì§¸ ë„í˜•(ì›)ì˜ ë„“ì´: {area2}')
else:
  print('ì˜ëª»ëœ ë„í˜• ì¢…ë¥˜ë¥¼ ì…ë ¥í–ˆìŠµë‹ˆë‹¤.')

# ë‘ ë„í˜•ì˜ ë„“ì´ ë¹„êµ
if area1 > area2:
  print(f'ì²«ë²ˆì§¸ ë„í˜•({shape_type1})ì´ {area1-area2}ë§Œí¼ ë” ë„“ìŠµë‹ˆë‹¤.')
elif area1 < area2:
  print(f'ë‘ë²ˆì§¸ ë„í˜•({shape_type2})ì´ {area2-area1}ë§Œí¼ ë” ë„“ìŠµë‹ˆë‹¤.')
else:
  print('ë‘ ë„í˜•ì˜ ë„“ì´ê°€ ê°™ìŠµë‹ˆë‹¤.')
```

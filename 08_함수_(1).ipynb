### 함수의 scope
* 전역 변수: 모듈(파일) 내에서 접근할 수 있으며 모듈 내 어디에서든지 접근할 수 있다.
* 지역 변수: 함수 내에서 정의된 변수이며, 해당 함수 내에서만 접근 할 수 있다. 해당 함수의 실행이 종료되면 더 이상 함수 안의 지역 변수에 접근할 수 없다.

````python
num = 41 # 전역변수
def number ():
  num = 3 # 지역변수
  print(f'함수 안에서, num={num}')

  number()
  print(f' global, num{num}')
````

### 재귀 함수 : 자기 자신을 호출하는 함수

````python
def factorial(n):
  if n == 1:
    return 1
  else:
    return n * factorial(n-1)

print(factorial(3))
````

### 가변 인자

*args : 순차적인 입력에 대해서 개수에 상관없이 입력을 받을 수 있다.
```

```

**kwargs : 키워드 인자에 대해서 어떤 이름의 인자 입력이든 자유롭게 입력 받을 수 있다.
```

```

func(*args, **kwargs) : 함수 호출시 어떤 형태의 입력이든 받을 수 있다.

* 시간 복잡도 : 입력의 크기에 따라 알고리즘이 실행되는 데에 필요한 시간의 양을 나타내는 개념. 빅오 표기법(Big O Notation)으로 표시,


```
O(1) 입력의 크기 n에 대해 관계없이 동일한 시간이 걸리는 복잡도 표현
O(n) 입력의 크기 n에 대해 선형적인 시간이 걸리는 복잡도 표현
O(nlogn) 입력의 크기 n에 대해 n*logn에 비례하는 시간이 걸리는 복잡도 표현 (log밑2)

O(n**2) 입력의 크기 n의 제곱에 비례하는 시간이 걸리는 복잡도 표현

O(n**3) 입력의 크기 n의 세제곱에 비례하는 시간이 걸리는 복잡도 표현
O(a**n) 입력의 크기 n에 지수적인 관계로 시간이 걸리는 복잡도 표현 (a>1)
```

### ë¬¸ì œ: ì „ ì§ì›ì˜ ì›”ê¸‰ ì´ì•¡ êµ¬í•˜ê¸°

### ğŸ¯ ìˆ«ì ë§íˆê¸° ê²Œì„
1. **ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * `random.randint(1, 100)`ìœ¼ë¡œ ì •ë‹µì„ ìƒì„±í•©ë‹ˆë‹¤.
    * ì‚¬ìš©ìê°€ ì •ë‹µì„ ë§í ë•Œê¹Œì§€ `while` ë¬¸ìœ¼ë¡œ ì…ë ¥ì„ ë°˜ë³µí•©ë‹ˆë‹¤.
    * ì…ë ¥ê°’ì´ í¬ë©´ "ë„ˆë¬´ ë†’ì•„ìš”", ì‘ìœ¼ë©´ "ë„ˆë¬´ ë‚®ì•„ìš”"ë¥¼ ì¶œë ¥í•˜ì—¬ ë²”ìœ„ë¥¼ ì¢í˜€ê°‘ë‹ˆë‹¤.
2. **ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * ì‚¬ìš©ìê°€ ìˆ«ìë¥¼ í•˜ë‚˜ ì •í•˜ë©´ ì»´í“¨í„°ê°€ `low`ì™€ `high` ë³€ìˆ˜ë¥¼ ì¡°ì •í•˜ë©° ì •ë‹µì„ ì°¾ì•„ê°€ëŠ” ë¡œì§ì…ë‹ˆë‹¤.

```python
# 1. ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ«ì ë§ì¶”ê¸°
import random

print('1ë¶€í„° 100ê¹Œì§€ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í–ˆìŠµë‹ˆë‹¤.')
print('ì´ ìˆ«ìë¥¼ ë§ì¶°ë³´ì‡¼.')

answer = random.randint(1,100) #1~100ì‚¬ì´ ìˆ«ì í•˜ë‚˜ ëœì ìœ¼ë¡œ ì €ì¥
time = 0 #ë„ì „ íšŸìˆ˜

while True :
  user = int(input('ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: '))
  time += 1

  if user > answer:
    print(f'{user}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤!')

  elif user < answer:
    print(f'{user}ë³´ë‹¤ í½ë‹ˆë‹¤!')

  else :
    print(f'ì •ë‹µì…ë‹ˆë‹¤! {time}ë²ˆ ë§Œì— ë§ì·„ì—‰!')
    break
```

```python
# 2.ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ë¥¼ ì»´í“¨í„°ê°€ ë§ì¶”ê¸°
print('1ë¶€í„° 100ê¹Œì§€ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í•˜ì„¸ìš”.')
print('ê·¸ ìˆ«ìë¥¼ ì œê°€ ë§ì¶°ë³´ê² ìŠµë‹ˆë‹¤.')

low = 1
hight = 100
times = 0

while True :
  num = (low+hight)//2
  answer = input(f'{num}ì¸ê°€ìš”? ok / up / down')
  times += 1

  if answer == 'ok':
    print(f'ê·¸ëŸ¼ ì •ë‹µì€{answer}ì´ì•¼!. ë‚´ê°€ {times}ë²ˆë§Œì— ë§ì·„ì–´!')
    break
  elif answer == 'up':
    low = num+1
  elif answer == 'down':
    hight = num-1
  else :
    print('ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”. (ok / up / down)')
```

### ğŸ¤– ë²ˆì—­ê¸°
1. **ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜**: êµ¬ê¸€ ë²ˆì—­ ê¸°ëŠ¥ì„ ì‚¬ìš©í•˜ê¸° ìœ„í•´ `googletrans` ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì„¤ì¹˜í•©ë‹ˆë‹¤.
2. **ë²ˆì—­ ë¡œì§**:
    * `Translator()` í•¨ìˆ˜ë¡œ ë²ˆì—­ê¸° ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
    * `translator.translate(ë¬¸ì¥, dest='ì–¸ì–´ì½”ë“œ')`ë¥¼ ì‚¬ìš©í•˜ì—¬ ë²ˆì—­í•©ë‹ˆë‹¤.
    * 'en'ì€ ì˜ì–´, 'ja'ëŠ” ì¼ë³¸ì–´ ë“±ì„ ì˜ë¯¸í•˜ë©° ë‹¤ì–‘í•œ êµ­ê°€ì˜ ì–¸ì–´ë¡œ ë³€í™˜ì´ ê°€ëŠ¥í•©ë‹ˆë‹¤.

```python
import googletrans
translator  = googletrans.Translator()
while True:
  lang = input("ë²ˆì—­í•  ì–¸ì–´ë¥¼ ì„ íƒí•˜ì„¸ìš”!(ì˜ì–´/ì¼ë³¸ì–´/ë…ì¼ì–´/ì¢…ë£Œ) ")
  if lang == "ì˜ì–´":
    lang = "en"
  elif lang == "ì¼ë³¸ì–´":
    lang ="ja"
  elif lang == "ë…ì¼ì–´":
    lang ="de"
  elif lang == "ì¢…ë£Œ":
    print("ì¢…ë£Œ")
    break
  else:
    print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤.")
    continue
  src = input("ë²ˆì—­í•  ë‚´ìš©ì„ ì…ë ¥í•˜ì„¸ìš”! ")
  dest = translator.translate(src, dest=lang)
  print(dest.text)
```

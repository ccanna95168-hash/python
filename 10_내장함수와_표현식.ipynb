### key 입력 사용


```
max(입력, key=함수)
max(입력, key=lambda x: x.내장함수())
```



````python
nums = [1,-3,5,4]
print(sorted(nums,key=abs)) # abs()함수 : 절대값
````

````python
name = ['apple','kiwi','banana']
print(sorted(name,key=len)) # len()함수 : 길이로 변환
````

````python
names = ['BANANA','GRAPE','apple']
print(sorted(names)) # 사전 순서로 정렬, 대문자/소문자/0~9 숫자
print(sorted(names,key=str.lower)) # lower() 함수 :  소문자로 변환
````

````python
names = [[5,6],[-30,-10,-10]]
print(max(names, key=len))
print(min(names, key=len))
````

### zip() 함수 : 둘 이상의 반복 가능한 값을 입력 받아 각각의 요소를 순서대로 묶어서 튜플 형태로 반환하는 파이선 내장 함수

````python
nums = [1,2,3]
names = ['one','two','three']
for item in zip(nums,names):
  print(item)
````

````python
nums = [1,2,3]
names = ['one','two','three']
num_strs='123'
for item in zip(nums,names,num_strs):
  print(item)
````

````python
nums = [1,2,3]
names = ['one','two']
for item in zip(nums,names):
  print(item)
````

### enumerate() 함수 :  여러 요소로 이루어진 값을 입력 받아 인덱스 값과 요소 값이 담긴 튜플 형태로 제공하는 파이썬 내장 함수

````python
names = ['banana','grape','apple']
for i, name in enumerate(names,1): #1은 시작값 (없애면 0인덱스부터 시작)
  print(i,name)
````

### if문 확장
* all(), any()를 사용한 if문
* short circuit evaluation : 먼저 등장한 논리식이 뒤의 논리식과 관계없이 전체 연산 결과를 결정할 경우 뒤의 논리식을 실행하지 않는다


```

  실행코드
```

* 시간 복잡도 : 입력의 크기에 따라 알고리즘이 실행되는 데에 필요한 시간의 양을 나타내는 개념. 빅오 표기법(Big O Notation)으로 표시,


```
O(1) 입력의 크기 n에 대해 관계없이 동일한 시간이 걸리는 복잡도 표현
O(n) 입력의 크기 n에 대해 선형적인 시간이 걸리는 복잡도 표현
O(nlogn) 입력의 크기 n에 대해 n*logn에 비례하는 시간이 걸리는 복잡도 표현 (log밑2)

O(n**2) 입력의 크기 n의 제곱에 비례하는 시간이 걸리는 복잡도 표현

O(n**3) 입력의 크기 n의 세제곱에 비례하는 시간이 걸리는 복잡도 표현
O(a**n) 입력의 크기 n에 지수적인 관계로 시간이 걸리는 복잡도 표현 (a>1)
```
